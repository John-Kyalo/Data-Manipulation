CONCAT DATAFRAMES

```python
# create dummy data
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
 

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                     'B': ['B4', 'B5', 'B6', 'B7'],
                     'C': ['C4', 'C5', 'C6', 'C7'],
                     'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])
 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                     'B': ['B8', 'B9', 'B10', 'B11'],
                     'C': ['C8', 'C9', 'C10', 'C11'],
                     'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
```

```python
# combine dataframes
result = pd.concat([df1, df2, df3])
result
```

```python
# combine dataframes
result = pd.concat([df1, df2, df3], keys=['x', 'y', 'z'])
result
```

```python
# get second dataframe
result.loc['y']
```

```python
#concat d4 and d1(outer join)
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                        'D': ['D2', 'D3', 'D6', 'D7'],
                        'F': ['F2', 'F3', 'F6', 'F7']},
                       index=[2, 3, 6, 7])
    

result = pd.concat([df1, df4], axis=1, sort=False)
result
```

```python
#concat d4 and d1 with an inner join
result = pd.concat([df1, df4], axis=1, join='inner')
result
```

MERGING DATAFRAMES

```python
# create dummy data
df_a = pd.DataFrame({
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']})

df_b = pd.DataFrame({
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']})

df_c = pd.DataFrame({
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]})
```

```python
#merge dfa and dfc based on subjectid
pd.merge(df_a, df_c, on='subject_id')
```

SQL JOINS

Outer join (return all records, if no match- NULL

```python
pd.merge(df_a,df_b, on ='subject_id', how='outer')
```

INNER JOIN(ONLY MATCHING VALUES IN BOTH TABLES)

```python
pd.merge(df_a, df_b, on='subject_id', how='inner')
```

RIGHT JOIN(all records from TABLE B,with the matching in TABLE A, IF NO MATCH-NULL ON LEFT SIDE)

```python
pd.merge(df_a, df_b, on='subject_id', how='right')
```

LEFT JOIN (all records from TABLE A, with the matching in TABLE B, IF NO MATCH -NULL ON RIGHT SIDE)

```python
pd.merge(df_a, df_b, on='subject_id', how='left')
```

MERGE OR CONCAT?

One major difference is that merge() is used to combine dataframes on the basis of values of common columns. Whileconcat() is used to append dataframes one below the other (or sideways, depending on whether the axis option is set to 0 or 1).
